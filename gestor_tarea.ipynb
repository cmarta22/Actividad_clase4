agregar al menos 3 tareas distintas al programa utilizando el metodo insert ()de la lista. cada tarea debe ser presentada como diccionario con la clave nombre,descripcion y estado.

```python
tareas=[]
tarea1={"nombre":"carlos","descripcion":"limpiar dormitorio","estado":"pendiente"},
tareas2={"nombre":"raul","descripcion":"ordenar la cosina","estado":"terminado"},
tareas3={"nombre":"juan","descripcion":"limpiar el jardin","estado":"en proseso"}
print(tareas)
```

```python
tareas.insert(0,tarea1)
tareas.insert(1,tareas2)
tareas.insert(2,tareas3)
```
